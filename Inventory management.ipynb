## Product Details
1. Product ID 
2. Name
3. Price
4. Quantity

```python
with open("Inventory.txt", 'r') as fd:
    products = fd.read().splitlines()

    
print("Current Inventory:")
print("------------------------------")
for prod in products:
    prod_details = prod.split(",")
    print(f"ID: {prod_details[0]}, Name: {prod_details[1]}, Price: {prod_details[2]}, Quantity: {prod_details[3]}")
print("------------------------------\n")


from datetime import datetime

try:
    #taking user input
    ui_name = input("Please type your name: ")
    ui_phone = int(input("Please enter your phone number: "))
    ui_email = input("Please enter your email id: ")
    ui_prod_id = input("Enter Product ID: ")
    ui_prod_qn = input("Enter Product Quantity: ")
    updated_prod_list = []
    # Validate quantity input
    if not ui_prod_qn.isdigit():
        raise ValueError("Error: Quantity must be a positive number")
    
    ui_prod_qn = int(ui_prod_qn)
    if ui_prod_qn <= 0:
        raise ValueError("Error: Quantity must be greater than 0")
    
    product_found = False
    
    for prod in products:
        prod_details = prod.split(",")
        
        if prod_details[0] == ui_prod_id:
            product_found = True
            available_stock = int(prod_details[3])
            
            # Check if requested quantity exceeds available stock
            if ui_prod_qn > available_stock:
                raise ValueError(f"Error: Requested quantity ({ui_prod_qn}) exceeds available stock ({available_stock})")
            elif available_stock <= 0:
                raise ValueError("Error: Product is out of stock")
            
            print("------------------------------")
            print("Product Name:   ", prod_details[1])
            print("Price:          $" + str(int(prod_details[2])))
            print("Quantity:       ", ui_prod_qn)
            print("------------------------------")
            print("Billing Amount: $" + str(ui_prod_qn * int(prod_details[2])))
            print("------------------------------")
            
            prod_details[3] = str(available_stock - ui_prod_qn)
            
            #creating sales file
            try:
                transaction_time = datetime.now().strftime("%d/%m/%Y %I:%M %p")  # 03/11/2024 02:30 PM
                with open("sales.txt", 'a') as sales_fd:
                    purchase_amount = ui_prod_qn * int(prod_details[2])
                    sales_detail = f"{ui_name},{ui_phone},{ui_email},{ui_prod_id},{ui_prod_qn},${purchase_amount}, {transaction_time}\n"
                    sales_fd.write(sales_detail)
            except Exception as e:
                print(f"Warning: Could not record sale: {str(e)}")
                
        updated_prod_list.append(",".join(prod_details))
    
    if not product_found:
        raise ValueError(f"Error: Product ID {ui_prod_id} not found")

except ValueError as e:
    print(str(e))
except Exception as e:
    print(f"An error occurred: {str(e)}")

with open("Inventory.txt", 'w') as fd:
    fd.write('\n'.join(updated_prod_list))
```
